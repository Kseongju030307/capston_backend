```python


model = YOLO(r"F:\종합설계\yolo_for_classification\dataset\runs\detect\train4\weights\best.pt") 


results = model.val()
```

```python
import numpy as np
metrics = results.box  
class_names = results.names  

# Precision, Recall, mAP 데이터 가져오기
precision_vals = np.array(metrics.p) if hasattr(metrics, 'p') else np.array([0])
recall_vals = np.array(metrics.r) if hasattr(metrics, 'r') else np.array([0])
map50_vals = np.array(metrics.map50) if hasattr(metrics, 'map50') else np.array([0])
map50_95_vals = np.array(metrics.map) if hasattr(metrics, 'map') else np.array([0])

# 최대/최소 값 및 해당 클래스 인덱스 찾기
precision_max_idx, precision_min_idx = precision_vals.argmax(), precision_vals.argmin()
recall_max_idx, recall_min_idx = recall_vals.argmax(), recall_vals.argmin()
map50_max_idx, map50_min_idx = map50_vals.argmax(), map50_vals.argmin()
map50_95_max_idx, map50_95_min_idx = map50_95_vals.argmax(), map50_95_vals.argmin()

# 최대/최소 값의 클래스 이름 가져오기
precision_max_label = class_names.get(precision_max_idx, "Unknown")
precision_min_label = class_names.get(precision_min_idx, "Unknown")
recall_max_label = class_names.get(recall_max_idx, "Unknown")
recall_min_label = class_names.get(recall_min_idx, "Unknown")
map50_max_label = class_names.get(map50_max_idx, "Unknown")
map50_min_label = class_names.get(map50_min_idx, "Unknown")
map50_95_max_label = class_names.get(map50_95_max_idx, "Unknown")
map50_95_min_label = class_names.get(map50_95_min_idx, "Unknown")

# 평균, 최대, 최소 계산
precision_mean = precision_vals.mean()
recall_mean = recall_vals.mean()
map50_mean = map50_vals.mean()
map50_95_mean = map50_95_vals.mean()

# 결과 출력
print(f"Precision (P-score) - 평균: {precision_mean:.4f}, 최대: {precision_vals.max():.4f} ({precision_max_label}), 최소: {precision_vals.min():.4f} ({precision_min_label})")
print(f"Recall (R-score) - 평균: {recall_mean:.4f}, 최대: {recall_vals.max():.4f} ({recall_max_label}), 최소: {recall_vals.min():.4f} ({recall_min_label})")
print(f"mAP@0.5 : {map50_mean:.4f}")
print(f"mAP@0.5:0.95 : {map50_95_mean:.4f}")
```
